```python
import numpy as np
import matplotlib.pyplot as plt
```

## 正常柱状图 .bar

```python
x = np.array(['a', 'b', 'c', 'd'])
y = np.array([4, 8, 5, 12])

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111)

ax.bar(x = x, height = y)

ax.set_title('Test BarBar')
ax.set_xlabel('category')
ax.set_ylabel('values')

plt.show()
```

## 水平柱状图 .hbar

```python
x = np.array(['a', 'b', 'c', 'd'])
y = np.array([4, 8, 5, 12])

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111)

ax.barh(y = x, width = y)

ax.set_title('Test BarBar')
ax.set_xlabel('values')
ax.set_ylabel('category')

plt.show()
```
