```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 3, 3, 4, 6, 8, 9])
y = np.array([0, 3, 5, 5, 7, 7, 8])

fig = plt.figure(figsize = (6, 6))
ax = fig.add_subplot(111)
ax.grid(True)

ax.scatter(x = x, y = y, marker = 'o', edgecolors = 'black', label = 'test', s = 100)

###
for i, txt in enumerate(y):
    plt.text(x = x[i] + 0.2, y = y[i], s = txt, fontsize = 15)
###

ax.legend(loc = 'lower right', shadow = True)

plt.show()
```
