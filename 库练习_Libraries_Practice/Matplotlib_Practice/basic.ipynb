## 创建画布 - 创建子图 - 选定子图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
'''
两种不同的画图方式: plt. & ax. 
'''

x, y = np.arange(1, 5), np.arange(1, 5)

##### 第一种 plt #####

# 生成一个画布
fig = plt.figure(figsize = (4, 4))

# 在此画布上隐式的生成一个画图区域来作图
plt.plot(x, y)

plt.show()
```

```python
##### 第二种 ax #####

# 先生成一个画布
fig = plt.figure(figsize = (4, 4))

# 在此画布上选定一个子区域作了一个‘子图’
ax = fig.add_subplot(111) # 分成1x1的部分，并对第1个子图进行操作
ax.plot(x, y)

plt.show()
```

```python
'''

在 Matplotlib 中，ax 通常用作表示子图（subplot）的变量名。"ax" 是 
"axes" 的缩写，代表一个坐标轴对象。在 Matplotlib 中，子图实际上就是
一组坐标轴，每个坐标轴可以绘制自己的图形。

使用 ax 的好处在于，它提供了更灵活和面向对象的方式来创建和管理图形。
通过显式地创建 Figure 对象和 Axes 对象（即子图的坐标轴），你可以更精
确地控制图形的外观和布局。

'''
x1, y1 = np.array([1, 2, 3, 5]), np.array([1, 3, 4, 5])
x2, y2 = np.array([1, 2, 2, 5]), np.array([1, 3, 4, 5])

fig = plt.figure(figsize = (4, 4))

#                   行 列 位置
ax1 = fig.add_subplot(221)
ax1.plot(x1, y1)

ax2 = fig.add_subplot(222)
ax2.plot(x1, y2)

ax3 = fig.add_subplot(223)
ax3.plot(x2, y1)

ax4 = fig.add_subplot(224)
ax4.plot(x2, y1)

plt.show()
```

## 修改/添加各项属性（添加标题/下，xy轴名字，修改xy轴刻度与范围）

```python
# 背景：水果店苹果销售记录

apple = np.array([78, 80, 79, 81, 91, 95, 96])
day = np.arange(1, 8)

fig = plt.figure(figsize = (6, 6))
ax = fig.add_subplot(111)

ax.plot(day, apple)

plt.show()
```

```python
'''
修改
'''

apple = np.array([78, 80, 79, 81, 91, 95, 96])
banana = np.array([70, 80, 81, 82, 75, 90, 89])
day = np.arange(1, 8)

fig = plt.figure(figsize = (6, 6))
ax = fig.add_subplot(111)

#########

# 设置图线风格
ax.plot(day, apple, 'r-.d', label = 'Apple')

ax.plot(day, banana, 'c-d', label = 'Banana')

#########

################################################

# 设置刻度范围
ax.set_xlim(0.9, 7.1)
ax.set_ylim(40, 100)

# 设置显示的刻度
ax.set_xticks(np.linspace(1, 7, 7)) # 生成等差数列
ax.set_yticks(np.linspace(50, 100, 6))

# 设置刻度的标签
ax.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
                   fontsize = 12, rotation = 20)
ax.set_yticklabels(['50kg', '60kg', '70kg', '80kg', '90kg', '100kg'],
                   fontsize = 12)

# 设置xy轴的标签
ax.set_xlabel('Day')
ax.set_ylabel('Sales volume')

# 设置标题
ax.set_title('Apple Sales', fontsize = 18, color = 'red')

# 添加图例
ax.legend(loc = 'lower right', shadow = True)

################################################

plt.show()
```
