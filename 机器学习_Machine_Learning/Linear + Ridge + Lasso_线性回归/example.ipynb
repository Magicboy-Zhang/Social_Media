```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
x_data = np.array([4, 6, 7]).reshape(-1, 1)
y_data = np.array([3, 4, 8])
xx = np.arange(2, 9, 0.1).reshape(-1, 1)

reg = LinearRegression()
knn = KNeighborsRegressor(n_neighbors = 1)

y_pred = reg.fit(x_data, y_data).predict(xx)
y_predk = knn.fit(x_data, y_data).predict(xx)
```

```python
fig = plt.figure(figsize = (6, 6))
ax = fig.add_subplot(111)
ax.grid(True)

ax.scatter(x = x_data, y = y_data, edgecolors = 'black')
ax.plot(xx, y_pred, 'r', label = 'Linear Regression')
ax.plot(xx, y_predk, 'b', label = '1-NN')

ax.set_xlim(2, 9)
ax.set_ylim(2, 9)

ax.legend(loc = 'lower right', shadow = True)

plt.show()
```
